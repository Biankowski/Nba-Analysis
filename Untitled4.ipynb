```python
# This function is responseble to convert the Height that is in feet, to Meters
def height_to_meters(height_i):
    int_part, remainder = divmod(height_i, 1)
    foot_to_meters = int_part / 3.281
    formated_ramainder = format(remainder, '.2f')
    if formated_ramainder == '0.10' or formated_ramainder == '0.11':
        inches_to_cm = (float(formated_ramainder) * 100) / 39.37
    else:
        inches_to_cm = (float(formated_ramainder) * 10) / 39.37

    height_in_meters = foot_to_meters + inches_to_cm

    return round(height_in_meters, 2)
```

```python
# Create a new column in the dataset and apply the convert function to it
df['Height_meters'] = df['Height_i'].apply(height_to_meters)
```
